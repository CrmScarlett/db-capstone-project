```python
import mysql.connector as connector
import mysql.connector.errors as errors
from mysql.connector.pooling import MySQLConnectionPool

# Task1: Connect client with your database and interact with the data it holds
## Storeing database connection information in a dictionary
dbconfig = {"database":"littlelemondb"
            , "user":"root" #replace with your_user_name
            , "password":"Test!123"} #replace with your_password

## Creating a connection pool with two connections
pool_b = MySQLConnectionPool(pool_name = "pool_b", pool_size = 2, **dbconfig)
print("The connection pool is created with a name: ", pool_b.pool_name)

# Implement a try-except block to handle any possible errors
try:
    connection1 = pool_b.get_connection()
    print("Connection is successfully established.")
except connector.Error as er:
    print("Error code", er.errno)
    print("Error message", er.msg)

cursor = connection1.cursor()

# Task2: Execute a query to show the tables in the database
show_tables_query = "SHOW tables" 
cursor.execute(show_tables_query)

result = cursor.fetchall()
print("The tables in the database are: ")
print(result)


# Task 3: Query with table JOIN and filtering
query = """ 
 SELECT O.OrderID
,O.Quantity * M.Price AS TotalCost
,CONCAT(C.FirstName, " ", C.LastName) AS CustomerName
,C.PhoneNumber
,CONCAT(C.Address,", ",C.City,", ",C.State) AS Address
FROM Orders O
INNER JOIN Menu M ON O.MenuID = M.MenuID
INNER JOIN Customers C ON C.CustomerID = O.CustomerID
WHERE C.CustomerID <> 1
AND O.Quantity * M.Price > 60
ORDER BY TotalCost DESC"""

cursor.execute(query)
result1 = cursor.fetchall()
print("Customer consums more than $60: ")
for row in result1:
    print(row)

# Close the cursor and connection
cursor.close()
connection1.close()
```
